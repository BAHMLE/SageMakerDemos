Import required libraries

```python
from numpy import loadtxt
import torch
import torch.nn as nn
import torch.optim as optim
import pandas as pd
```

Load `pima-indians-diabetes.csv` dataset

```python
dataset_name = 'pima-indians-diabetes.csv'
bucket_data_name = 'bah-data'
data_location = 's3://{}/{}'.format(bucket_data_name, dataset_name)

data = pd.read_csv(data_location)
X = data.iloc[:, :len(data.columns)-1].values
y = data.iloc[:, len(data.columns)-1].values
```

```python
X = torch.tensor(X, dtype=torch.float32)
y = torch.tensor(y, dtype=torch.float32).reshape(-1, 1)
```

Define the model

```python
model = nn.Sequential(
    nn.Linear(8, 12),
    nn.ReLU(),
    nn.Linear(12, 8),
    nn.ReLU(),
    nn.Linear(8, 1),
    nn.Sigmoid()
)
```

```python
print(model)
```

Set hyperparameters

```python
loss_fn = nn.BCELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
n_epochs = 20
batch_size = 10
```

Training a model

To build a training loop use two nested for-loops, one for epochs and one for batches

```python
for epoch in range(n_epochs):
    for i in range(0, len(X), batch_size):
        Xbatch = X[i:i + batch_size]
        y_pred = model(Xbatch)
        ybatch = y[i:i + batch_size]
        loss = loss_fn(y_pred, ybatch)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
    print(f'Finished epoch {epoch}, latest loss {loss}')
```

Evaluation of the model

```python
with torch.no_grad():
    y_pred = model(X)
accuracy = (y_pred.round() == y).float().mean()
print(f"Accuracy {accuracy}")
```
